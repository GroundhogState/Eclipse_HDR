## Useful functions

```python
def cv_darkfield(dark_img, edges=[],visual=0,verbose=0):
    # Returns blue/green/red channel means ,stds, and histogram 
    # counts according to input edges
    [b, g, r] = [dark_img[:, :, 0], dark_img[:, :, 1], dark_img[:, :, 2]]
    mns = list(map(lambda x: np.mean(x), [b,g,r]))
    sts = list(map(lambda x: np.std(x), [b,g,r]))
    
    if len(edges)==0:
        edges = np.arange(0,np.amax(dark_img),300)
    hist_vals = np.zeros([3,len(edges)-1])
    for channel in range(3):
        [hist_vals[channel,:],_]=np.histogram(dark_img[:,:,channel],edges-0.5)
    centres=np.mean([edges[1:],edges[:-1]],axis=0)
    if verbose>0:
        print("Means: ", mns)
        print(" Stds: ", sts)
    if visual>0:
        plt.figure(figsize=(12, 12))
        plt.subplot(221)
        plt.plot(centres,hist_vals[0,:],color="blue")
        plt.plot(centres,hist_vals[1,:],color="green")
        plt.plot(centres,hist_vals[2,:],color="red")
        plt.ylim([0,1e6])
        plt.xlim([0,max(edges)])
        plt.subplot(222)
        plt.plot(centres,hist_vals[0,:],color="blue")
        plt.plot(centres,hist_vals[1,:],color="green")
        plt.plot(centres,hist_vals[2,:],color="red")
        plt.xlim([0,max(edges)])
        plt.ylim([1,1e6])
        plt.yscale('log')
        plt.tight_layout()
        plt.show()
    return [mns, sts, hist_vals]
```

## Noise statistics

```python
max_val = 13000
bin_size = 10
edges = np.r_[-0.5:max_val+.5:bin_size]
centres = np.r_[0:max_val:bin_size]

all_M = np.zeros([num_dark,3])
all_S = np.zeros([num_dark,3])
all_hist = np.zeros([num_dark,3,len(centres)])

for idx in tqdm_notebook(range(len(dark_paths))):
    # flag 2 for 16bit GRB images
#     img_ = cv2.imread(dark_paths[idx],3)
    [all_M[idx,:],all_S[idx,:],all_hist[idx,:,:]] = cv_darkfield(img_array[idx],edges)
```

```python
mean_hist = np.mean(all_hist,axis=0)
ch_mean = np.mean(all_M,axis=0)
ch_std = np.mean(all_S,axis=0)

full_mean = np.mean(all_M,axis=0)
full_std = np.mean(all_S,axis=0)


plt.figure()

plt.figure(figsize=(12,6))
plt.subplot(221)
plt.errorbar(dark_exp,all_M[:,0],all_S[:,0]/np.sqrt(num_pixels),color='Blue')
plt.errorbar(dark_exp,all_M[:,1],all_S[:,1]/np.sqrt(num_pixels),color='green')
plt.errorbar(dark_exp,all_M[:,2],all_S[:,2]/np.sqrt(num_pixels),color='r')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Colour time')
plt.title('Colour mean')

plt.subplot(222)
plt.scatter(all_M[:,0],(all_S[:,0]),color='Blue')
plt.scatter(all_M[:,1],(all_S[:,1]),color='green')
plt.scatter(all_M[:,2],(all_S[:,2]),color='r')
plt.scatter(full_mean,full_std,100,color='k')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Colour mean')
plt.ylabel('Colour variancee')
plt.title('Mean vs variance')

plt.subplot(223)
plt.plot(centres,mean_hist[0,:],c='b')
plt.plot(centres,mean_hist[1,:],c='g')
plt.plot(centres,mean_hist[2,:],c='r')
plt.yscale('log')

plt.tight_layout()
plt.show()

print('B noise: %.2f +- %.2f'%(ch_mean[0],ch_std[0]))
print('G noise: %.2f +- %.2f'%(ch_mean[1],ch_std[1]))
print('R noise: %.2f +- %.2f'%(ch_mean[2],ch_std[2]))
print(full_mean,full_std)
```

The signal is noisy, but it's clear that at least over this timescale, the noise is independent of the exposure time. This suggests that the noise is from the sensor readout, not the integration time. So the average of the darkfield images is, a *fairly* good approximation of the sensor noise (although the camera was probably hotter when I took the eclipse images, so the readout noise might be higher there). 

For the most part, there's a clear straight-line relationship between the mean and variance. For a Poissonian process, the mean equals the variance. This seems almost true here, with some deviations. The variance is also larger than the mean by about 100 with reasonable consistency - this could be due to the mean being 'dragged down' by the overabundance of zero-value pixels. A couple of oddly-behaved images buck the trend, with considerably larger variances than expected by the trend. These histograms appear somewhat 'stretched'.

## Sensor correction

```python
darkfield = np.load(out_root+"noise_1577598754.npy")
# Subtract off the noise floor
def sensor_correction(imgs,dark_img,img_labels,out_folder):
    timestamp = time.time()
    signal_support = np.zeros(num_imgs)
    for (idx,img) in tqdm_notebook(enumerate(imgs),'Denoising',len(imgs)):
        num_below_floor = np.sum(img_denoise<0)
        num_above_floor = np.sum(img_denoise>0)
        num_px = num_above_floor+num_below_floor
        signal_support[idx] = num_above_floor/num_px
        np.save(out_folder+"denoise_%u_"%timestamp+img_labels[idx][:-4],img_denoise)
        print('%u: %u/%u pixels below noise floor, %.2f pc recovery'%(idx,num_below_floor,num_px,num_above_floor/num_px))
    return signal_support

signal_support = sensor_correction(img_array,darkfield,img_fnames,denoise_out)
```

```python
# The signal support grows linearly with exposure, but sadly many exposures are quite short.
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.plot(signal_support)
plt.xlabel('Shot number')
plt.subplot(1,3,2)
plt.plot(exp_array,signal_support)
plt.xlabel('Shot exposure (s)')
plt.subplot(1,3,3)
plt.plot(exp_array,signal_support)
plt.xlabel('Shot exposure (s)')
# plt.xscale('log')
plt.show()
```

## Exposure correction

```python
pre_adjust_means = np.zeros(num_imgs)
post_adjust_means = np.zeros(num_imgs)
elist = []
pre_patch_average[idx] = np.zeros(num_imgs)
post_patch_average[idx] = np.zeros(num_imgs)


for (idx,fname) in tqdm_notebook(enumerate(img_fnames)):
    denoised_img = np.load(denoise_out+"denoise_%s.npy"%(fname[:-4]))
    exposure_corrected = denoised_img/exp_array[idx]
    # Try averaging the whole image to begin with, but probably worth restricting to patches
    M1 = denoised_img>0
    M2 = denoised_img<2**15
    elist.append(exp_array[idx])
    pre_adjust_means[idx] = np.mean(denoised_img[M1 & M2])
    post_adjust_means[idx] = np.mean(exposure_corrected[M1 & M2])
    pre_patch_average[idx] = np.mean(denoised_img)
    post_patch_average[idx] = np.mean(exposure_corrected)
    np.save(exposure_out+fname[:-4]+'.npy',exposure_corrected)
    
```
